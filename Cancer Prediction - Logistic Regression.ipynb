## Breast Cancer prediction using Logistic Regression

Breast cancer is the most common cancer amongst women in the world. It accounts for 25% of all cancer cases, and affected over 2.1 Million people in 2015 alone. It starts when cells in the breast begin to grow out of control. These cells usually form tumors that can be seen via X-ray or felt as lumps in the breast area.

The key challenges against it’s detection is how to classify tumors into malignant (cancerous) or benign(non cancerous).

This Dataset is obtained from Breast Cancer Wisconsin. It contains information of cells and its various features. Source of dataset https://www.kaggle.com/datasets/yasserh/breast-cancer-dataset. 

**We will predict whether the cell is malignant or benign.** The Dependant Variable is binary (categorical) so we will be using **Logistic Regression** and Python libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Breast Cancer Data.csv')
```

```python
df.head()
```

**The Dependant variable here is ' diagnosis ' which is either M or B**

```python
df.shape
```

**Logistic Regression - We will feed numerical data into our model, we will run logistic regression and the output should be a Categorical value M or B, Yes or No**

```python
df.info()
```

#### Looking at the summary Statistics of the dataset (numerical variables)

```python
df.describe()
```

#### Checking for Missing Values

```python
df.isnull().sum()
```

**Unnamed: 32 has all fields missing. We can drop this column. Additionally we'll drop the id column as we dont need this information.**

```python
df.drop(['Unnamed: 32','id'], axis=1, inplace=True)
```

#### Checking for Duplicates

```python
df[df.duplicated()==True]
```

No duplicates

```python
df.head()
```

#### Converting ' diagnosis ' into 1 and 0 (binary variable)

```python
df.diagnosis = [1 if value == 'M' else 0 for value in df.diagnosis]
```

```python
df.head()
```

## Training Logistic Regression Model

#### The target variable and predictors

```python
y = df['diagnosis']
X = df.drop(['diagnosis'], axis=1)
```

#### Normalising the data

Some of the numerical variable values are less than 1 while others are big numbers above 1. So to deal with such huge variation in magnitude of numbers we normalize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Create a Scalar object
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
X_scaled = scaler.fit_transform(X)
X_scaled
```

#### Spliting the Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.3, random_state=42)
```

#### Training the Model

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()

regressor.fit(X_train, y_train)

# Predit the target variable on test data

y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

#### Comparing the predicted values (y_pred) with actual values of y (y_test)

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)
print(f"The Accuracy was: {round(accuracy*100)} %")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```
